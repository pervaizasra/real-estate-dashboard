ðŸ§¹ Step 3: Clean Column Names

```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
```

Step 4: Handle Missing Values

```python
df['agency'].fillna('Unknown', inplace=True)
df['agent'].fillna('Unknown', inplace=True)
```

ðŸ“… Step 5: Convert Data Types

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.drop('area', axis=1, inplace=True)
```

```python
df['property_type'] = df['property_type'].str.title().str.strip()
df['purpose'] = df['purpose'].str.title().str.strip()
df['location'] = df['location'].str.title().str.strip()
```

```python
df['province_name'] = df['province_name'].replace({
    'Islamabad Capital': 'Islamabad',
    'Sindh ': 'Sindh'  # trailing space fix example
})
```

ðŸš« Step 8: Remove Outliers


```python
df = df[df['price'] < 1e8]  # remove unrealistic prices over 100M
df = df[df['area_size'] > 0]  # ensure area is positive
```
